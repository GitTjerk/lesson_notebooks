```python
def save_all_years(years_list):
    for year in years_list:
        vehicles = data[data['Year'] == year]
        if vehicles.shape[0] < 1200:
            vehicles.to_csv(f'{year}_vehicles.csv')
        else:
            print(f'Could not save year {year} because it is too large')
```
