```python
import pandas as pd
```

```python
obs_ages = pd.read_csv('ages_population.csv')
```

```python
obs_ages.head()
```

```python
obs_ages[['observation']].plot(kind='bar', figsize=(50,30))
```

```python
obs_ages[['observation']].sort_values(by='observation', ascending=False).plot(kind='bar', figsize=(50,30))
```

```python
freq = obs_ages.observation.value_counts()
```

```python
frequencies_df = pd.DataFrame(data = freq).sort_index().reset_index().rename(columns={'index': 'age'})
```

```python
frequencies_df
```

```python
frequencies_df[['observation']].plot(kind='bar', figsize=(50,30))
```

Not true because we don't have any observation with 3 years we need to define another dataframe with all the possible ages

```python
poss_ages = pd.DataFrame(data = {'age': pd.Series(range(100))})
```

```python
poss_ages.merge(frequencies_df, how='left')
```

```python
new_freq = poss_ages.merge(frequencies_df, how='left').fillna(0)
```

```python
new_freq[['observation']].plot(kind='bar', figsize=(50,30))
```
