```python
import random

MY_DICE_TODAY = 10

def roll_the_dice(stop = MY_DICE_TODAY):
    return random.randrange(1, stop)

def roll_the_dice_many_times(stop = MY_DICE_TODAY, times = 1):
    results = []
    for i in range(times):
        results.append(roll_the_dice(stop))
    return results
```

```python
simulation = roll_the_dice_many_times(times = 100)
```
