```python
import random
```

```python
MY_DICE_TODAY = 10

def roll_the_dice(stop = MY_DICE_TODAY):
    dice_result = random.randrange(1, stop)
    print(f"You got a {dice_result}!")
    
    
def roll_the_dice_many_times(stop = MY_DICE_TODAY, times = 1):
    for i in range(times):
        roll_the_dice(stop)
```

```python
roll_the_dice_many_times()
```

```python
roll_the_dice_many_times(times = 1000)
```

We need to save the result. For that, we need a value, not only the print


```python
##
def roll_the_dice(stop = MY_DICE_TODAY):
    dice_result = random.randrange(1, stop)
    return dice_result
##

def roll_the_dice(stop = MY_DICE_TODAY):
    return random.randrange(1, stop)

def roll_the_dice_many_times(stop = MY_DICE_TODAY, times = 1):
    results = []
    for i in range(times):
        results.append(roll_the_dice(stop))
    return results
```

```python
roll_the_dice_many_times(times = 100)
```

```python
simulation = roll_the_dice_many_times(times = 100)
```
